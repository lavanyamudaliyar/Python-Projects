```python


names = ['group1', 'group2', 'group3']
values = [1, 10, 100]


plt.figure(figsize=(10,7))

plt.subplot(231)# total Row, total columns, plot no.
plt.bar(names,values)

plt.subplot(232)
plt.scatter(names,values)

plt.subplot(233)
plt.plot(names,values)

plt.subplot(234)# total Row, total columns, plot no.
plt.bar(names,values)

plt.subplot(235)
plt.scatter(names,values)

plt.subplot(236)
plt.plot(names,values,'r-', linewidth=10) #Linewidth
#1-9, A-F colors combination
plt.show()
```

```python
names = ['group1', 'group2', 'group3']
values = [1, 10, 100]


plt.figure(figsize=(10,5))

plt.subplot(131)# total Row, total columns, plot column
plt.bar(names,values,color="red")
plt.title('Plot 1')

plt.subplot(132)
plt.scatter(names,values)
plt.title('Plot 2')

plt.subplot(133)
plt.plot(names,values, 'r-*')
plt.title('plot 3')

plt.suptitle('subplots') #super title
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])

plt.bar(x,y,color='black', width= 0.5) #width for normal bar
plt.title('Bar Graphs')
plt.show()
```

# Creating pie chart

```python
y = np.array([15,35,15,25])
plt.figure(figsize=(20,10))
mylabels=["a1","a2","a3","a4"]
plt.pie(y, colors = colors, labels=mylabels) 

plt.show()
```

# Explode

```python
y = np.array([15,35,15,25])
mylabels=["a1","a2","a3","a4"]
myexplode=[0.1,0.4,0.2,0.3]
plt.pie(y, colors = colors, labels=mylabels, explode=myexplode) 
plt.show()
```

# Shadow

```python
y = np.array([15,35,15,25])
mylabels=["a1","a2","a3","a4"]
myexplode=[0.1,0.4,0.2,0.3]
plt.pie(y, colors = colors, labels=mylabels, explode=myexplode, shadow= True) #1-9, A-F colors combination
plt.show()
```

# Legend

```python
y = np.array([15,35,15,25])
mylabels=["a1","a2","a3","a4"]
plt.pie(y, colors = colors, labels=mylabels)

plt.legend()
plt.show()
```
